```python
def teclado(comando, digitos):
    if comando.isdigit():
        digitos.append(comando)
    elif comando == 'C':
        if len(digitos) > 0:
            digitos.pop()
    elif comando.startswith('D'):
        k = int(comando[2:])
        if k <= len(digitos):
          del digitos[-k:]
    elif comando.startswith('M'):
        i, j = map(int, comando[2:].split())
        if i>=1 and j<= len(digitos):
            print(''.join(digitos[max(i-1, 0):j]))

digitos = list()
while True:
    comando = input().strip()
    if comando == 'end':
        break
    teclado(comando, digitos)
```

```python
jugadores_adentro=list()
while True:
  num_jugador=int(input())
  if num_jugador==0:
    break
  elif num_jugador in jugadores_adentro:
    jugadores_adentro.pop(jugadores_adentro.index(num_jugador))
  elif (num_jugador+1) in jugadores_adentro:
    jugadores_adentro.pop(jugadores_adentro.index(num_jugador+1))
  elif (num_jugador-1) in jugadores_adentro:
    jugadores_adentro.pop(jugadores_adentro.index(num_jugador-1))
  else:
    jugadores_adentro.append(num_jugador)
if len(jugadores_adentro)==0:
  print(0)
else:
  for i in jugadores_adentro:
    print(i,end=" ")
```

```python
import bisect
import math

M = int(input())
nums = list()

while True:
    num = int(input())
    if num == 0:
        break
    nums.append(num)
    if len(nums) % M == 0:
        nums.sort()
        mid = len(nums) // 2
        if len(nums) % 2 == 0:
            num1, den1 = nums[mid - 1], 1
            num2, den2 = nums[mid], 1
            median_num = num1 + num2
            median_den = 2
        else:
            median_num = nums[mid]
            median_den = 1
        
        gcd = math.gcd(median_num, median_den)
        median_num //= gcd
        median_den //= gcd
        
        if median_den == 1:
            print(median_num)
        else:
            print("{}/{}".format(median_num, median_den))
```
