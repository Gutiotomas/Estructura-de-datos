```python
from collections import deque 
NT=tuple(map(int,input().split()))
RK=deque() 
for i in range(NT[0]):
  RK.append(tuple(map(int,input().split())))
N=NT[0]
T=NT[1]
C=0
while len(RK)>=1 and (T- RK[0][1])>0:
  if C==4:
    T-=RK.popleft()[1]
    C=0
  else:
    T-=RK[0][1]
    RK.append(RK.popleft())
    C+=1
if len(RK)==0:
  print("quedaron boletas disponibles")
else:
  print(RK[0][0],T)
```

```python
from collections import deque
fila=deque()
while True:
    instruccion = input()
    if instruccion.startswith("agrega "):
        numero = int(instruccion.split()[1])
        fila.append(numero)
    elif instruccion == "engulle":
        if len(fila) > 1:
            if fila[0] > fila[-1]:
                fila.pop()
            elif fila[-1]>=fila[0]:
                fila.popleft()
        elif len(fila) == 1:
            fila.pop()
    elif instruccion == "termina":
        break

if len(fila) == 1:
    print(f"cabeza {fila[0]} cola {fila[0]}")
elif fila:
    print(f"cabeza {fila[0]} cola {fila[-1]}")
else:
    print("uroboro vacio")
```

```python
def balanceado(expresion):
    pila = []
    abrir = "([{"
    cerrar = ")]}"
    for c in expresion:
        if c in abrir:
            pila.append(c)
        elif c in cerrar:
            if not pila:
                return False
            elif abrir.index(pila.pop()) != cerrar.index(c):
                return False
    return not pila

n = int(input())
for i in range(n):
    expresion = input()
    if balanceado(expresion):
        print("correcta")
    else:
        print("incorrecta")
```

```python
from collections import deque

for i in range (int(input())):
  posicion=tuple(map(int,input().split()))[1]
  cola=deque(map(int,input().split()))
  tiempo=0
  elfo=0
  while True:
    tiempo+=5
    elfo+=1
    cola[0]-=1
    if cola[0]==0:
      if elfo==posicion:
        break
      else:
        if elfo<posicion:
          elfo-=1
          posicion-=1
          cola.popleft()
        else:
          elfo-=1
          cola.popleft()
    else:
      cola.append(cola.popleft())
    if elfo==len(cola):
      elfo=0
  print(tiempo)
```

```python
for i in range(int(input())):
    num_discos = int(input())
    varilla_a = list(range(num_discos, 0, -1))
    varilla_b = []
    varilla_c = []
    movimientos_validos = True
    while True:
        movimiento = input().split()
        if movimiento[0] == 'X':
            break
        origen = movimiento[0]
        destino = movimiento[1]
        if origen == destino:
            continue
        if origen == 'A':
            varilla_origen = varilla_a
        elif origen == 'B':
            varilla_origen = varilla_b
        else:
            varilla_origen = varilla_c
        if destino == 'A':
            varilla_destino = varilla_a
        elif destino == 'B':
            varilla_destino = varilla_b
        else:
            varilla_destino = varilla_c
        if not varilla_origen:
            movimientos_validos = False
            continue
        if varilla_destino and varilla_destino[-1] < varilla_origen[-1]:
            movimientos_validos = False
            continue
        disco = varilla_origen.pop()
        varilla_destino.append(disco)
    solucion_valida = False
    if movimientos_validos and varilla_c == list(range(num_discos, 0, -1)):
        solucion_valida = True
    if not movimientos_validos:
        print("secuencia invalida")
    elif solucion_valida:
        print("soluciona la torre")
    else:
        print("no soluciona la torre")
```
