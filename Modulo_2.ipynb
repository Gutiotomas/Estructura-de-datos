```python
x=int(input())
for i in range(x):
    y=input()
    y=y.split()
    y.reverse()
    for i in range(len(y)):
        if i%2==0 and y[i]!=y[-1]:
            y[i],y[i+1]=y[i+1],y[i]
    y="".join(y)
    print(y)
```

```python
N=int(input())
SO = tuple(int(i) for i in str(input()).split(", "))
LAR = tuple(int(i) for i in str(input()).split(", "))
IS = tuple(int(i) for i in str(input()).split(", "))
X=0
Y=0
Z=0
for i in range (N):
  if (SO[i]+LAR[i]+IS[i])%2==SO[i]%2:
    X+=1
  if (SO[i]+LAR[i]+IS[i])%2==LAR[i]%2:
    Y+=1
  if (SO[i]+LAR[i]+IS[i])%2==IS[i]%2:
    Z+=1
print("SO:", X, ", LAR:", Y, ", IS:", Z, sep="")
```

```python
N=int(input())
for i in range(N):
  M=tuple(str(i) for i in str(input()).split(" "))
  Mul=M[::-1]
  Mpot=str()
  for i in range (len(Mul)):
    if i%2==1:
      Mpot+=Mul[i]
      Mpot+=Mul[i-1]
  if len(Mul)%2==1:
    Mpot+=Mul[-1]
  print(Mpot)
```

```python
N = int(input())
Ld= tuple(str(i) for i in str(input()).split(", "))
Lr= Ld[::-1]
word=str()
for i in range(N//2):
  word+=Ld[i];word+=Lr[i]
if len(Ld)%2==1:
  word+=Ld[(N//2)]
print(word)
```
