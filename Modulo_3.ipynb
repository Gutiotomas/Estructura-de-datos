```python
def div(numero):
  return [i for i in range(1, numero+1) if numero % i == 0]
for i in range(int(input())):
  T, N = tuple(map(int,input().split()))
  num = [int(i) for i in input().split(" ")] #leer los numeros O(N)
  divisores = div(int(N)) #O(P)
  divisor = 0
  for i in range(int(T)): #O(N)
    if num[i] == divisores[divisor]:
      divisor += 1
  if len(divisores) == divisor: print("Es PrimiConjunto")
  else: print("No es PrimiConjunto")
```

```python
def Bin(X, e):
  ini = 0
  fin = len(X)-1
  while ini <= fin:
    med = (ini + fin)//2
    if X[med] == e: return med
    elif e < X[med]: fin = med - 1
    else: ini = med + 1
  return -1
A, C, N = int(input()),[int(i) for i in input().split(" ")],int(input())
C.sort()
for i in range(N):
  x,y = input().split(" ")
  P = abs(Bin(C, int(x))-Bin(C, int(y)))
  print(f"{P} kms")
```

```python
def bubbleSort(X):
  N = len(X)
  aux = 0
  for i in range(1, N):
    for j in range(N-i):
      if X[j] > X[j+1]:
        X[j], X[j+1] = X[j+1], X[j]
        aux += 1
  return aux
for i in range(int(input())):
  C = [int(i) for i in input().split()]
  print(bubbleSort(C))
```

```python
for i in range(int(input())):
  P = tuple(sorted(map(int,input().split(", ")),reverse = True))
  SI = P[0]
  SD = sum(P[1:])
  D = abs(SI-SD)
  for i in range(1,len(P)):
    SI,SD = SI + P[i],SD - P[i]
    DA = abs(SI-SD)
    if D < DA:
      break
    else: D = DA
  print(D)
```
