```python
from ast import If
import heapq
A = list()
B = list()
C = list()
puntosA = 0
puntosB = 0
puntosC = 0
salenA = 0
salenB = 0
salenC = 0

while True:
    comandos = input().split()
    if comandos[0] == "A":
      heapq.heappush(A, int(comandos[1]))
    elif comandos[0] == "B":
      heapq.heappush(B, int(comandos[1]))
    elif comandos[0] == "C":
      heapq.heappush(C, int(comandos[1]))
    elif comandos[0] == "menores":
      if len(A)>0:
        salenA=heapq.heappop(A)
      else: 
        salenA=1001
      if len(B)>0:
        salenB=heapq.heappop(B)
      else:
        salenB=1001
      if len(C)>0:
        salenC=heapq.heappop(C)
      else:
        salenC=1001
      if salenA==1001 and salenB==1001 and salenC==1001:
        continue
      if min([salenA,salenB,salenC]) == salenA:
        puntosA+=1
      if min([salenA,salenB,salenC]) == salenB:
        puntosB+=1
      if min([salenA,salenB,salenC]) == salenC:
        puntosC+=1      
    else:
      break
print("Equipo A:", puntosA)
print("Equipo B:", puntosB)
print("Equipo C:", puntosC)
```
